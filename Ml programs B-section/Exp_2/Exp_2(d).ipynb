```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
'Day': [1, 2, 3, 4, 5, 6],
'Temperature': [22, 24, 23, 26, 28, 30]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.plot(x='Day', y='Temperature', kind='line', legend=True)
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")
plt.title("Line Plot of Temperature Over Days")
# Show plot
plt.show()
```
