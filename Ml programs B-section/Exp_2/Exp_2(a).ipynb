```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
 'Maths': 85,
 'Physics': 90,
 'Chemistry': 78,
 'Biology': 88
}
```

```python
marks = pd.Series(data, name="Marks")
```

```python
marks.plot(kind='bar', legend=True)
```

```python
plt.title("Marks in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```
